```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dog_breeds.csv")

print(df.head())
```

```python
#Puuduvad väärtused ja andmetüüpide kontroll
print(df.info())
print("\nPuuduvad väärtused:")
print(df.isnull().sum())
```

```python
#Duplikaatide kontroll
print("Duplikaate:", df.duplicated().sum())
```

```python
df_unique = df.drop_duplicates(subset=["Breed"])
```

```python
print(df_unique[df_unique["Breed"] == "Golden Retriever"])
```

```python
print(df[["Height (in)", "Longevity (yrs)"]].describe())
```

```python
print(df["Height (in)"].head(10))

print(df["Longevity (yrs)"].head(10))
```

```python
# Numbriliste veergude muutmine numbriteks
def range_to_mean(value):
    low, high = value.split("-")
    return (float(low) + float(high)) / 2
```

```python

df["Height (in)"] = df["Height (in)"].apply(range_to_mean)
df["Longevity (yrs)"] = df["Longevity (yrs)"].apply(range_to_mean)
```

```python
print(df.dtypes)
print(df["Height (in)"].head())
print(df["Longevity (yrs)"].head())
```

```python

#Inchies to cm
df["Height (cm)"] = df["Height (in)"] * 2.54
```

```python
sns.scatterplot(

    data=df,

    x="Height (cm)",

    y="Longevity (yrs)"

)

plt.xlabel("Kõrgus (cm)")

plt.ylabel("Eluiga (aastates)")

plt.title("Koera kõrguse ja eluea seos")

plt.show()
```
